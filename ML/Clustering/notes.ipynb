#### 1. K-Means Clustering
##### It is a centroid-based clustering algorithm. It partitions the data into 'K' distinct, non-overlapping clusters.
##### It is also useful to work with above 2D data.
##### K value could be even bcx in this k denotes clusters

#### Steps: 
##### 1- Decide n clusters.
##### 2- Initiallize centriods.
##### 3- Assign Cluster.
##### 4- Move Centroids.
##### 5- Finish

##### The low distance data with newly centroids goes into that cluster.
##### Suppose k=3, it randomly select 3 data as centriods.
##### To move centroids after first cluster formed, we calculate the mean of features which are on x-axis and y-axis and then move to center.
##### If current and previous centroids are equal then we stop otherwise we continue.

##### Elbow MEthod: It tell us how many cluster possible to make.

#### Usage: 
##### 1- Customer Segmentation.
##### 2- Document clustering.

#### Drawbacks: 
##### Need to specify number of cluster before start.
##### Different initializations can lead to different final clusters.
##### Outliers can significantly skew(Direction or position change) the position of centroids.

#### Solution:
##### DBSCAN or Mean Shift can handle non-spherical shapes and varying densities also handle outliers data.


#### 2. DBSCAN
##### It is a density-based algorithm. It groups together points that are closely packed (points with many nearby neighbors) and marks points that lie alone in low-density regions as outliers or noise. 

#### MinPoints and Epsilon 
##### eps (Ïµ): The maximum distance between two samples.
##### min_samples: The number of samples in a neighborhood for a point to be considered as a core point. 

#### Core Points, Border Points and Noise Points
##### Core Points: Data is equal or greater than min_samples inside epsilon radius
##### Border Points: Data which has low data compared to min_samples but at least one core points.
##### Noise Points: If no border or core points.

#### Density Connected Points:
##### If 2 points are density connected then add in one cluster.
##### If all pattern have same distance or d < epsilon, it means density connected.
##### If distance > epsilon then it can't be connected.

#### Steps: 
##### 0- min_samples = 3, eps = 2(distance)
##### 1- Identify all points as core, border or noise points.
##### 2- For all of the unclustered points.
#####   a- Create new cluster.
#####   b- Add all the points that are unclustered and density connected to the current point into the cluster.
##### 3- For each unclustered border point assign itto the cluster of nearest core points.
##### 4- Leave all the noise points as it is.

#### Advantage: 
##### Robust to outliers. Built in capability to odentify outliers.
##### No need to specify clusters.
##### Can find arbitrary shaped clusters.
##### Only 2 hyperparameters to tune.

#### Usage: 
##### Clustering data with irregular shapes.
##### Geospatial data analysis.


#### Drawbacks: 
##### Performance can be sensitive to the choice of eps and min_samples, which can be hard to determine, especially if clusters have varying densities.
