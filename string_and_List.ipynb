```python
#13. Reverse Each Word in a Sentence

def reverse_string(str):
  splitted = str.split()
  reversed_words = []
  for i in splitted:
    reversed = i[::-1]
    reversed_words = ''.join(reversed)
    result = ''.join(reversed_words)
    print(result)

reverse_string('hello world')
reverse_string('Python is fun')
```

```python
# 14. Find the Longest Common Prefix of a List of Strings

def longest_prefix(strs):
    prefix = strs[0]

    for word in strs[1:]:
        while not word.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    print(prefix)

input1 = ["flower", "flow", "flight"]
input2 = ["dog", "racecar", "car"]
longest_prefix(input1)
longest_prefix(input2)

```

```python
# 16. Find the Second Most Frequent Character in a String
def frequent_char(val):
  frequency = {}

  for char in val:
      if char in frequency:
          frequency[char] += 1
      else:
          frequency[char] = 1

  sorted_chars = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

  if len(sorted_chars) < 2:
      return None

  return sorted_chars[1][0]

frequent_char('banana')
frequent_char('mississippi')
```

```python
#18. Find the Longest Palindromic Substring

def longest_palindromic(s):
  def expand_around_center(left, right):
      while left >= 0 and right < len(s) and s[left] == s[right]:
          left -= 1
          right += 1
      return s[left+1:right]

  longest = ""
  for i in range(len(s)):
      temp = expand_around_center(i, i)
      if len(temp) > len(longest):
          longest = temp

      temp = expand_around_center(i, i+1)
      if len(temp) > len(longest):
          longest = temp

  return longest

print(longest_palindromic('babad'))  # Output: 'bab' or 'aba'
print(longest_palindromic('cbbd'))  # Output: 'bb'

print(longest_palindromic('babad'))
print(longest_palindromic('cbbd'))
```

```python
# 22. Implement Run-Length Encoding

def length_encoding(value):
    result = ""
    count = 1

    for char in range(1, len(value)):
        if value[char] == value[char - 1]:
            count += 1
        else:
            result += str(count) + value[char - 1]
            count = 1
    result += str(count) + value[-1]
    return result


print(length_encoding('aaabbc'))
print(length_encoding('wwwwaaadexxxxxx'))
```

```python
# 23. Find All Permutations of a String

def find_permutations(s):
  if len(s) == 0:
      return [""]

  permutations = []
  for i in range(len(s)):
      char = s[i]
      remaining = s[:i] + s[i+1:]
      for p in find_permutations(remaining):
          permutations.append(char + p)

  return permutations




print(find_permutations("abc"))
print(find_permutations("ab"))
```

```python
# 16. Find Pairs in a List That Sum to a Target Value
def trageted_sum(val, sum):
  paris = []
  for i in range(len(val)):
    for j in range(i + 1, len(val)):
      if val[i] + val[j] == sum:
        paris.append((val[i], val[j]))
  return paris

print(trageted_sum([1, 2, 3, 4, 5] , 6))
print(trageted_sum([2, 4, 6, 8] , 10))

```

```python
# 17. Find the Product of All Elements Except Self

def find_product(val):
  result = []
  for i in range(len(val)):
    product = 1
    for j in range(len(val)):
      if i != j:
          product *= val[j]
    result.append(product)
  return result


print(find_product([1, 2, 3, 4]))
print(find_product([5, 6, 7]))
```

```python
# 18. Find the Longest Consecutive Sequence in a List
def longest_consecutive_sequence(nums):
    num_set = set(nums)
    longest = 0

    for num in num_set:
        if num - 1 not in num_set:
            length = 1
            current = num

            while current + 1 in num_set:
                current += 1
                length += 1

            longest = max(longest, length)

    return longest

print(longest_consecutive_sequence([100, 4, 200, 1, 3, 2]))
print(longest_consecutive_sequence([10, 11, 12, 50, 51, 52]))
```

```python
# 22. Partition a List into Even and Odd Elements

def find_even_odd(val):
  even = []
  odd = []
  for num in val:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)
  return [even] + [odd]

print(find_even_odd([1, 2, 3, 4, 5]))
print(find_even_odd([10, 15, 20, 25]))
```

```python
# 23. Find the First Non-Repeating Element in a List

def largest_non_repeating(val):
    frequency = {}

    for num in val:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1

    for num in val:
        if frequency[num] == 1:
            return num

    return None


print(largest_non_repeating([4, 5, 1, 2, 0, 4]))
print(largest_non_repeating([10, 20, 10, 30, 20]))
```

```python
# 24. Sort a List Without Using Built-in Sort
def sort_list(val):

  for num1 in range(len(val)):
    for num2 in range( num1 + 1, len(val)):
      if val[num1] > val[num2]:
        val[num1] , val[num2] = val[num2] , val[num1]
  return val

print(sort_list([3, 1, 4, 1, 5, 9]))
print(sort_list([10, 5, 2, 8]))
```

```python
# 25. Find All Triplets That Sum to Zero

def targeted_triplets(nums, target):
    result = []
    n = len(nums)
    for num1 in range(n):
        for num2 in range(num1 + 1, n):
            for num3 in range(num2 + 1, n):
                if nums[num1] + nums[num2] + nums[num3] == target:
                    triplet = sorted([nums[num1], nums[num2], nums[num3]])
                    if triplet not in result:
                        result.append(triplet)
    return result

print(targeted_triplets([-1, 0, 1, 2, -1, -4], 0))
print(targeted_triplets([1, -2, 1, 0, 5], 0))
```
